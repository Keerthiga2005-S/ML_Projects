```python
import pandas as pd
```

```python
d=pd.read_csv("/content/weatherAUS - weatherAUS.csv")
```

```python
d.isna().sum()
```

```python
d
```

```python
d.shape
```

```python
d=d.drop('MinTemp',axis=1)
```

```python
d=d.drop('MaxTemp',axis=1)
```

```python
d=d.drop('WindSpeed9am',axis=1)
```

```python
d=d.drop('Temp9am',axis=1)
```

```python
for i in d.columns:
  if d[i].dtypes=='float64':
       d[i]=d[i].fillna(d[i].mean())
  if d[i].dtypes=='object':
       d[i]=d[i].fillna(d[i].mode()[0])
```

```python
d
```

```python
d.isna().sum()
```

```python
cate=[]
for col in d:
  if d[col].dtypes=='object':
    cate.append(col)
    print(col)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cate:
  d[i]=le.fit_transform(d[i])
```

```python
d
```

```python
x=d.drop('RainTomorrow',axis=1)
y=d['RainTomorrow']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,ytest))
```

```python
from sklearn.naive_bayes import BernoulliNB
model1=BernoulliNB()

model1.fit(xtrain,ytrain)
```

```python
y_pred1=model1.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred1,ytest))
```
