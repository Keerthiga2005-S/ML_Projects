```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
de=['customerID','gender','Partner','Dependents','InternetService','PhoneService','MultipleLines','OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','Churn','OnlineBackup']
for col in de:
  d[col]=le.fit_transform(d[col])
d
```

```python
d=d.replace(" ",0)
```

```python
x=d.iloc[:,0:20]
x
```

```python
y=d.iloc[:,-1:]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=8)
model.fit(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_pred,ytest))
print(confusion_matrix(y_pred,ytest))
```
